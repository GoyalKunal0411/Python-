### SCATTER PLOT

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
# Make a scatter plot between sepal length and sepal width
```

```python
plt.scatter(iris['sepal_length'],iris['sepal_width'],color='darkgreen',
           marker='*',s=35)
plt.show()
```

```python
# Make a multiple scatter plot between sepal lenght sepal width, sepal length an pela length
plt.scatter(iris['sepal_length'],iris['sepal_width'],color='darkgreen',
           marker='*',s=35)
plt.scatter(iris['sepal_length'],iris['petal_length'],color='darkred',
           marker='+',s=35)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [5,5]
```

```python

plt.figure(figsize=(5,3))
plt.suptitle('IRIS')  # for super title
plt.subplot(1,2,1)
plt.scatter(iris['sepal_length'],iris['sepal_width'],color='darkgreen',
           marker='*',s=35)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('SEPAL vs SEPAL')
plt.subplot(1,2,2)
plt.scatter(iris['sepal_length'],iris['petal_length'],color='darkred',
           marker='+',s=35)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('SEPAL vs PETAL')
plt.tight_layout()
plt.show()
```

#### BAR PLOT

```python
amount = [2500,3700,3000,4200,2200]
cust = ['Ajay','Suresh','Kamla','Neha','Vijay']
```

```python
# make a vertical barplot b/w amount and cust
plt.bar(x=cust,height=amount,
        width=0.7,color=['red','blue','green','yellow','orange'],
       edgecolor='k',lw=2,ls='--',align='edge')
t = 0
for i in amount:
    plt.text(t-0.25,i+70,i)
    t+=1
plt.tight_layout()
plt.show()
```

```python
# make a horizontal bar plot b/w amount and cust
plt.barh(y=cust,width=amount,height=0.7,color=['c'])
plt.show()
```

#### PIE PLOT

```python
countries = ['USA','France','Germany','India','Russia']
count = [63.76,34.54,45.11,57.43,60.54]
```

```python
# Make a pie plot
plt.pie(x=count,labels=countries,counterclock=False,startangle=90,
       autopct='%.0f%%',labeldistance=1.2,pctdistance=0.8,radius=0.8,
       shadow=True,explode=[0,0,0.1,0.1,0],colors=['r','b','g','c','m'])
plt.show()
```

```python
# make a doughnut chart
plt.pie(x=count,labels=countries,counterclock=False,startangle=90,
       autopct='%.0f%%',labeldistance=1.2,pctdistance=0.8,radius=0.8,
       shadow=True,explode=[0,0,0,0.1,0],colors=['r','b','g','c','m'])
plt.pie(x=[1],colors=['w'],radius=0.3)
plt.show()
```

### SEABORN

#### STRIP PLOT

```python
data = sns.load_dataset('titanic')
```

```python
# Make a strip plot between age and sex
```

```python
sns.stripplot(x=data['sex'],y=data['age'])
plt.show()
```

```python
data
```

```python
# Make a strip plot between age and sex and survived
sns.stripplot(x=data['sex'],y=data['age'],hue=data['survived'])
plt.show()
```

#### VIOLIN PLOT

```python
# Make a violin plot between age and sex
sns.violinplot(x=data['sex'],y=data['age'])
plt.show()
```

```python
# Make a violin plot between age and sex and survived
sns.violinplot(x=data['sex'],y=data['age'],hue=data['survived'])
plt.show()
```

#### SWARM PLOT

```python
# Make a swarm plot between age and sex
sns.swarmplot(x=data['sex'],y=data['age'])
plt.show()
```

```python
# Make a swarm plot between age and sex and survived
sns.swarmplot(x=data['embarked'],y=data['age'],hue=data['alive'])
plt.show()
```

#### PAIR PLOT

```python
# MAke a pair plot for iris dataset
sns.pairplot(x_vars=['sepal_length','sepal_width'],
             y_vars=['petal_length','petal_width'],
            data = iris,hue='species',
             palette={'setosa':'red','versicolor':'violet','virginica':'green'})
plt.show()
```

#### DISTRIBUTION PLOT

```python
# Make a distribution plot of Fare
sns.distplot(data['fare'])
plt.show()
```

```python
sns.displot(iris['sepal_length'],kind='kde')
```

#### COUNT PLOT

```python
# AMke a count plot of people sex in titanic
sns.countplot(x='sex',data=data)
```

#### REGRESSION PLOT

```python
# make a regression plot b/w sepal length and petal length
sns.regplot(iris['sepal_length'],iris['petal_length'],ci=0,
           line_kws={'color':'red','ls':'--'},color='green',marker='*')
```
